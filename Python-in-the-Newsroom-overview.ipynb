## Python has an extensive library of "modules," a collection of scripts people have developed to make things a little easier. I'm using a module called Pandas, a data analysis module. First we import it, and then I'll read in a simple spreadsheet file (.csv format).

```python
import pandas as pd
```

```python
# Use has tags to make comments that Python won't try to run as code. 
# I'm reading in a file from the California Civic Data Coalition with current ballot measures.
# Python doesn't save everything you read into it automatically, so we'll save the file as a temporary "variable"
# and we'll use that variable to explore this database further.
ballots=pd.read_csv("BallotMeasures.csv")
ballots
```

## Looks a lot like any spreadsheet you'd open in Excel or Google Sheets, doesn't it? 
## If we scroll down, we see that Python isn't displaying all 172 rows. That's fine, because we don't need to see them all at this point. 
## We can see from this view this data is a collection of referendum's and ballot measures in past elections in California, and contains general information about each item.
## Let's look at the general makeup of the data with Pandas' .info() function.

```python
ballots.info()
```

## This cell shows us most of the cells in these columns are words (objects, strings), while the calaccess_measure_id column is full of whole numbers (int64, integers), and one column has no data but is classified as decimal numbers (Float64). 
## This might not seem like much, but it tells me a great deal of what I should expect from this data. 
## I'm going to be asking a lot more "what are" type questions than I am "how much" type questions.
## Of course, most people want to know the quantity of things. It's important, and this is where Python gets easier than excel.
## For example, think about how you would go about finding out how many of a certain type of ballot measure was in this data. 
## Before you take more than two seconds, let's do it with Pandas' .value_counts() function.


```python
ballots.classification.value_counts()
```

## Okay, so how many different names are there for these (do we have any repeating ballot issues)?

```python
ballots.name.value_counts()
```

## Unfortunatlely, all of the names are unique. So, surely it'll take a lont time to figure out how many names might share a common word or phrase, right?
## If this were Excel, you could apply a filter to the name column, search a phrase, select the ones you want, then save that as a new sheet to preserve the original sheet.
## We can filter with Python though.

```python
ballots[ballots.name.str.contains("TERM LIMIT")]
```

## Nice, so there have been two term limit ballot iniatives in the past 20 years? Cool. 
## Not quite. 

```python
ballots[ballots.name.str.contains("TERM|LIMIT")]
```

## That little bar in the filter is called a pipe or pipette (SHIFT+\), and it's what's known as an "operator" in this function.
## So we can now see all the ballot items with the words "TERM" and "LIMIT" in them.
## That's great and all, but there's not much substance. 
## At this point, your Excel workbook might have two sheets and you now want to go back to the original question of how many classifications for each of these results. 
## In Python, we can just string commands together. 

```python
ballots[ballots.name.str.contains("TERM|LIMITS")].classification.value_counts()
```

## Let me break down what you're reading in this line of code.
## First, you're telling Python to take your "ballots" variable, open it up and look in the name column for your search terms. After that, take those results and look in the classification column, then count the number for each.
## That's not all. We can use the .groupby function to add "when?" to our question.


```python
ballots[ballots.name.str.contains("TERM|LIMITS")].groupby('election_date').classification.value_counts()
```

## Okay now let's do something even more fun. Let's do some visulalization on the fly.
## Let's import another library called matplotlib and then tell it to display comands in side of this Jupyter Notebook (that's what this display environment is called. There are many environments out there, this is just my favorite).

```python
import matplotlib
%matplotlib inline
```

```python
ballots[ballots.name.str.contains("TERM|LIMITS")].groupby('election_date').classification.value_counts().plot.bar()
```

## We basically analyzed this data, double-checked the accuracy of the classifications, added a new "date" parameter and then made a chart in a few minutes. The more familiar you get with these tools, that last line of code might be your first step. Depending on the machine you're working on, you might be able to create this chart faster than you could open the file in Excel (yes, I've tested that on my old work laptop and it was true for me). 

# Aggregate Data vs. Micro Data
## There's two main categories of datasets available to you as a reporter: aggregate and micro data.
## Aggregate data is more common and includes things like the U.S. Census. You should think of it like this: aggregate data is the compiled data from a survey. You know the population demographics, but you don't have every individual survey at hand.
## Micro Data would be the individual surveys, and it's not as frequently available to you.
## That's fine. There's nothing wrong with aggregate data, you just have to have confidence in the source - and describe its potential flaws as clearly as possible. 
## It's rare, but I do have a project that uses micro data, and that's what we'll move onto in the next notebook.
